재귀 함수: 함수 안에 자신의 함수를 다시 호출하는 함수

예시) 팩토리얼-> 5!=5*4*3*2*1
n!=n*(n-1)*(n-2)*...*1

```python
# 반복문을 이용한 팩토리얼 구현 코드
# 곱하기 연산이 진행되기 때문에 output의 초기값은 1입니다. 더하기 연산이 진행되는 경우는 0으로 초기값을 설정

def oz_factorial(n):
    output=1
    for i in range(1,n+1):
        output *= i
    return output

n=int(input("구하고자 하는 팩토리얼의 수를 입력해주세요"))
print(f'{n}의 결과는 {oz_factorial(n)}입니다.')
```

재귀함수를 이용해서 팩토리얼을 구현하는 방식

oz_factorial(n)=n *factorial(n-1)
oz_factorial(0)=1   #곱하기에 영향을 주지 않으니까 1로 처리하자

5!
oz_factorial(5)=5 * oz_factorial(4)
               =5 * 4 * oz_factorial(3)
               =5 * 4 * 3 * oz_factorial(2)
               =5 * 4 * 3 * 2 * oz_factorial(1)
               =5 * 4 * 3 * 2 * 1 * oz_factorial(0)
               =5 * 4 * 3 * 2 * 1 * 1

```python
# 재귀함수를 이용해 팩토리얼 구현 코드
def oz_factorial(n):
    if n == 0:
        return 1
    else:
        return n * oz_factorial(n-1)
    
n=int(input("구하고자 하는 팩토리얼의 수를 입력해주세요"))
print(f'{n}의 결과는 {oz_factorial(n)}입니다.')
```

재귀함수는 for문이나 while문 대신 사아ㅛㅇ되기 떄문에 코드가 간결해지고 불필요한 변수를 만들지 않아도 된다.
다음의 단점
    1. Stack overflow(무한루프)가 발생할 수 있다. -> (꼬리 재귀로 문제해결 가능)
    2. 지속적인 함수 호출로 인해 함수 안에 변수를 계속해서 선언하게 되고 이로 인해 
       과도한 메모리 사용과 속도 저하가 발생할 수 있따. -> (메모리제이션으로 문제해결 가능)

그래서 쵝근까지 재귀함수 대신 for문으로 구현하는게 더 안정적으로 프로그래밍이 가능하다는 의견. 하지만 이런 문제도 코드적으로 해결이 가능
과도한 메모리 사용과 속도 저하를 발생시키는 요인들을 메모리제이션으로 해결해보는 과정 알아보자.

재귀함수의 문제점
피보나치 수열: 첫째 및 둘째 항이 1이며 그 뒤에 모든 항은 바로 앞 두 항의 합인 수열로 이루어져 있다.

규칙
    *시작은 한쌍의 토끼
    *두달 이상된 토끼는 번식을 할 수 있고, 번식이 가능한 달부터 매달 새끼를 한 쌍씩 낳을 수 있음

```python
# 피보나치 수열 함수 실행 계산되는 횟수를 파악
# 파이썬은 함수 내부에서 함수 밖에 있는 변수 사용할 수 없음
# 그래서 global 키워드 함수 이용해 함수 내부에서 밖에 있는 변수 사용할 수 있도록 만들기

count=0

def oz_fibo(n):
    print(f'피보나치 수열 {n}을 구하는 중입니다.')
    global count
    count += 1

    if n == 1:
        return 1
    if n ==2:
        return 1
    else:
        return oz_fibo(n-1) + oz_fibo(n-2)
    
n=int(input("구하고자 하는 피보나치 수열의 수를 입력해주세요"))
oz_fibo(n)
print(f'피보나치 수열 {n}을 구하기 위해 계산된 횟수는 {count}입니다.')
```

메로리제이션: 
    *컴푸터 프로그램이 동일한 계산을 반복적으로 해야할때, 이전에 계산한 값을 메모리에 저장하여 중복적인 계산을 반복적으로 해야할때,
    이전에 계산한 값을 메모리에 저장하여 중복적인 계산을 제거해 전체적인 실행 속도를 빠르게 해주는 기버으로 동적 계획법의 핵신이 되는 기술

코드 핵심 설명

    *초기값으로 설정했던 첫번째와 두번째 결과값을 memo 변수에 dict형 자료형을 넣어 저장
    * oz_fibo()함수 실행 시 memo 변수에 계산할 피보나치 수열의 결과값이 있는지 확인. 있다면 계산 안하고 이미 계산되어 저장된
    결과값을 이용한다.
    * 만일 결과값이 저장되어 있지 않다면 계산을 한다. 그리고 계산된 결과값은 ㄷ음 계산에서 활용할 수 있도록 memo에 저장
    

```python
memo = {
    1: 1,
    2: 1,
}

count=0

def oz_fibo(n):
    print(f'피보나치 수열 {n}을 구하는 중입니다.')
    global count
    count += 1

    if n in memo:
        return memo[n]
 
    else:
        output = oz_fibo(n-1) + oz_fibo(n-2)
        memo[n] =output
        return output
    
n=int(input("구하고자 하는 피보나치 수열의 수를 입력해주세요"))
oz_fibo(n)
print(f'피보나치 수열 {n}을 구하기 위해 계산된 횟수는 {count}입니다.')
```
