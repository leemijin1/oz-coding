[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


파이썬은 C언어와 다르게 Pointer가 존재하지 않는다. 
포인터란 프로그래밍 언어에서 다른 변수 또는 그 변수의 메모리 공간주소를 가리키는 변수를 말한다.
포인터가 가리키는 값을 가져오는 것을 역참조라고 한다.
파이썬에서 메모리 주소를 변수를 구별하는 용도로 사용하는데 변수의 메모리 주소는 id()를 이용해서도 구할 수 있다.
파이썬에서는 변수에 값을 할당하면 object를 생성해서 값을 저장한 후 변수는 해당 object의 메모리 주소를 의미하는 
id값을 가지게 된다. 여기서 변수는 object의 label정도로 생각하면 된다.

[문제2]

Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요

----

컴퓨터는 모든 수를 0과 1로 이루어진 2진수로 표현하기 떄문에 실수형끼리 연산에서 마주치면 오차가 발생한다. 10진수 0.1은 
이진법에서는 무한소수이기 때문에 이진 소수로 정확하게 표현할 수 없다. 무한 소수로 표현하게ㅏ 되면 메모리 범위를 벗어나므로 
유한한 비트수에 맞춰 근사값을 얻게 된다. 그래서 실제 수와 거의 같지만 정확하게 같지 않고 오차가 생겨나게 된다.

[문제3]

슬라이싱에 대해 간단히 설명해주세요

슬라이스는 무언가 일부를 잘라낸다는 뜻인다. 시퀀스 슬라이스도 말그대로 시퀀스 객체의 일부를 잘라낸다. 
[]안에 시작 인덱스와 끝 인덱스를 지정하면 해당 범위의 리스트를 잘라서 가져올 수 있다. 여기서 주의할 것은 
끝 인덱스는 가져오려는 범위에 포함하지 않는다.
